```
import numpy as np
import matplotlib.pyplot as plt 
```

```
#Declaracion de constantes

g=9.81
m=0.0001
Cw=0.5
p=1.225
r=0.003
A=np.pi*r**2
k=0.5*Cw*p*A
v0=0

x_temp=100
t_temp=0
while x_temp>0 :
  x_temp = 100- (m/k)*np.log(np.cosh(np.sqrt(k*g/m)*t_temp))
  t_temp+=0.001

t= np.linspace(0, t_temp, 100, endpoint = True) 
```

```
#Graficar posicion

x = 100- (m/k)*np.log(np.cosh(np.sqrt(k*g/m)*t))

plt.xlabel("Tiempo (s)")
plt.ylabel("Posición (m)")
plt.plot(t,x);
plt.title("\nPosición vrs. tiempo") 
plt.show
```

```
#Graficar velocidad

v = -np.sqrt(m*g/k)*np.tanh(np.sqrt(k*g/m)*t)

plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad (m/s)")
plt.plot(t,v);
plt.title("\nVelocidad vrs. tiempo") 
plt.show
```

```
#Graficar aceleracion

a = -g*(1/np.cosh(np.sqrt(k*g/m)*t))**2

plt.xlabel("Tiempo (s)")
plt.ylabel("Aceleración (m/s²)")
plt.plot(t,a);
plt.title("\nAceleración vrs. tiempo") 
plt.show
```
